# 4. Evolutionary simulation

Evolutionary simulation asks how the share of strategies changes when better-performing strategies become more common. This solution is useful for modelling adaptation without assuming fully rational actors.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')

payoff_matrix = pd.DataFrame(
    [[3,0,3,1.5],[5,1,1.2,3],[3,0.8,3,2],[3.5,1,2.5,2.25]],
    index=['cooperate','defect','tit_for_tat','random'],
    columns=['cooperate','defect','tit_for_tat','random']
)
payoff_matrix
```

```python
def replicator(initial, generations=60):
    shares = np.array(initial, dtype=float) / np.sum(initial)
    history = [shares.copy()]
    A = payoff_matrix.values
    for _ in range(generations):
        fitness = A @ shares
        avg = shares @ fitness
        shares = shares * fitness / avg
        shares = shares / shares.sum()
        history.append(shares.copy())
    return pd.DataFrame(history, columns=payoff_matrix.index).rename_axis('generation').reset_index()

evo = replicator([0.25,0.25,0.25,0.25])
evo.tail()
```

```python
evo.set_index('generation').plot(figsize=(8,4))
plt.ylabel('Population share')
plt.title('Evolutionary dynamics among strategies');
```
