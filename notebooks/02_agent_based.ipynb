# 2. Agent-based simulation

Agent-based simulation represents decision makers as agents with simple behavioural rules. Here we compare classic strategies: always cooperate, always defect, tit-for-tat, and random.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')

payoffs = {('C','C'):(3,3), ('C','D'):(0,5), ('D','C'):(5,0), ('D','D'):(1,1)}
def score(a,b): return payoffs[(a,b)]
```

```python
class Strategy:
    name = 'base'
    def move(self, own, other): raise NotImplementedError
class AlwaysCooperate(Strategy):
    name = 'always cooperate'
    def move(self, own, other): return 'C'
class AlwaysDefect(Strategy):
    name = 'always defect'
    def move(self, own, other): return 'D'
class TitForTat(Strategy):
    name = 'tit for tat'
    def move(self, own, other): return 'C' if not other else other[-1]
class RandomStrategy(Strategy):
    name = 'random'
    def move(self, own, other): return np.random.choice(['C','D'])

def match(s1, s2, rounds=50):
    h1, h2, p1, p2 = [], [], 0, 0
    for _ in range(rounds):
        a, b = s1.move(h1,h2), s2.move(h2,h1)
        x, y = score(a,b); h1.append(a); h2.append(b); p1 += x; p2 += y
    return p1, p2

strategies = [AlwaysCooperate(), AlwaysDefect(), TitForTat(), RandomStrategy()]
records=[]
for s1 in strategies:
    for s2 in strategies:
        p1,p2 = match(s1,s2)
        records.append({'row_strategy':s1.name,'column_strategy':s2.name,'row_payoff':p1,'column_payoff':p2})
df = pd.DataFrame(records)
df.pivot(index='row_strategy', columns='column_strategy', values='row_payoff')
```

```python
sns.heatmap(df.pivot(index='row_strategy', columns='column_strategy', values='row_payoff'), annot=True, fmt='.0f', cmap='viridis')
plt.title('Agent strategy tournament: row payoff');
```
