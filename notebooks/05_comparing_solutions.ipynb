# 5. Comparing the four solutions

Each simulation solution answers a different kind of question.

| Solution | Best for | Main limitation |
|---|---|---|
| Payoff matrix | Understanding the formal model | Too simple for heterogeneous behaviour |
| Agent-based | Explicit behavioural rules | Results depend on chosen rules |
| Monte Carlo | Uncertainty and robustness | Computational summaries can hide mechanisms |
| Evolutionary | Population-level adaptation | Fitness assumptions can be strong |

## Suggested exercises

1. Change the payoff values. When does cooperation become easier?
2. Add a new strategy, such as grim trigger.
3. Increase the Monte Carlo error rate. Which strategy is most robust?
4. Change the evolutionary starting population. Do you reach the same endpoint?
